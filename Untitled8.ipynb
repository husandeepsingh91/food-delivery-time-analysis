```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Order_ID': [1,2,3,4,5,6,7,8],
    'Restaurant': ['A','B','A','C','B','A','C','B'],
    'Delivery_Time': [30,45,25,50,40,35,60,55],
    'Distance_km': [5,8,3,10,7,6,12,9]
}

df = pd.DataFrame(data)
df
```

```python
df['Delivery_Time'].mean()
```

```python
df.groupby('Restaurant')['Delivery_Time'].mean()
```

```python
sns.barplot(x='Restaurant', y='Delivery_Time', data=df)
plt.title("Average Delivery Time by Restaurant")
plt.show()
```

```python
plt.scatter(df['Distance_km'], df['Delivery_Time'])
plt.xlabel("Distance (km)")
plt.ylabel("Delivery Time")
plt.title("Distance vs Delivery Time")
plt.show()
```
